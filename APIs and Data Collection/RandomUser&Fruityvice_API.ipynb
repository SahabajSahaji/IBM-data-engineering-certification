To generate a table with information about the users, we can write a function containing all desirable parameters. For example, name,gender,city,etc.The parameters will depend on the requirements of the test to be performed .We call the Get Methods,Listed at the beginning of this notebook.Then, we return pandas dataframe with the users.

```python
def get_users():
    users=[]
    for user in RandomUser.generate_users(20):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"city":user.get_city(),"State":user.get_state(),"Email":user.get_email(),"DOB":user.get_dob(),"Picture":user.get_picture()})
    return pd.DataFrame(users)    
```

Example 2:Fruityvice API
Another more common way to use APIs , is through requests library. The next lab requests and HTTp, wil contain more information about request.

```python
import requests
import json
```

We will obtain the fruityvice API data usint reqeusts.get("url") function. The data is in a json format

```python
headers={
    "User-Agent":"Mozilla/5.0"
}
data = requests.get("https://web.archive.org/web/20240929211114/https://fruityvice.com/api/fruit/all",headers=headers)
```

```python
results=json.loads(data.text)
```

```python
pd.DataFrame(results)
```

The result is in a nested json format. The 'nutrition'column contains multiple subcolumns,so the data needs to be flattened or normalized

```python
df2=pd.json_normalize(results)
df2
```

Let's see if we can extract some information fromthsi dataframe. Perhaps, we need to know the family and genues of a cherry.

```python
cherry=df2.loc[df2["name"]=="Cherry"]
(cherry.iloc[0]['family']),(cherry.iloc[0]['genus'])
```

Max nutritions

```python
cols=['nutritions.calories','nutritions.fat',	'nutritions.sugar',	'nutritions.carbohydrates',	'nutritions.protein']

#Create sum column
df2['total']=df2[cols].sum(axis=1) #axis=1 means row wise sum

#Get row with maximum total
df3=df2.loc[df2['total'].idxmax()]#idxmax() finds the indes of the row with the maximu value
print(df3)
```
